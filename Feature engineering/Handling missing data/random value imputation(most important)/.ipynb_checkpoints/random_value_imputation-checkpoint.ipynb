```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Random value imputation in numerical data:

```python
df=pd.read_csv('titanic.csv',usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
X=df.drop(columns=['Survived'])
y=df['Survived']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
#to show difference in data distribution and features, we are creating an entirely new column for our operation 
X_train['Age_imputed']=X_train['Age']
X_test['Age_imputed']=X_test['Age']
```

```python
X_train
```

```python
#checking where are missing values in Age_imputed column
X_train['Age_imputed'][X_train['Age_imputed'].isnull()]
```

```python
X_train['Age_imputed'][X_train['Age_imputed'].isnull()]=X_train['Age'].dropna().sample(X_train['Age'].isnull().sum()).values
X_test['Age_imputed'][X_test['Age_imputed'].isnull()]=X_test['Age'].dropna().sample(X_test['Age'].isnull().sum()).values
```

```python
X_train
```

```python
#checking data distribution before value imputation and after value imputation
sns.distplot(X_train['Age'],label='Original',hist=False)
sns.distplot(X_train['Age_imputed'],label = 'Imputed',hist=False)

plt.legend()
plt.show()
```

```python
print('Original variable variance: ', X_train['Age'].var())
print('Variance after random imputation: ', X_train['Age_imputed'].var())
```

```python
X_train[['Fare','Age','Age_imputed']].cov()
```

```python
#not much effect on outliers
X_train[['Age','Age_imputed']].boxplot()
```

```python
X_train['Age_imputed'] = X_train.apply(
    lambda obs: X_train['Age'].dropna().sample(1, random_state=int(obs['Fare'])).values[0]
    if pd.isnull(obs['Age']) else obs['Age'],
    axis=1
)
```
